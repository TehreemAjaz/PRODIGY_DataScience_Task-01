```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#load the data

data = sns.load_dataset('titanic')
```

```python
# print the data
data.head()
```

```python
data.info()
```

```python
# Plot the gender distribution
plt.figure(figsize=(8, 6))
sns.countplot(x='sex', data=data)
plt.title('Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```
